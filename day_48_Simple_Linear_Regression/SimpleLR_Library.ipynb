## Simple Linear Regression - Scikitlearn Linrary

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package(in LPA)')
```

```python
X = df.iloc[:,:1]
y = df.iloc[:,1:]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
X_test
```

```python
y_test
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
# print((X_test.iloc[0].values.reshape(1,1)))
# print(X_test)
lr.predict(X_test)
```

```python
plt.scatter(X,y, label='data points')
#plt.scatter(X_test,lr.predict(X_test), color='red', label = 'regression model')
#plt.plot(X_test,lr.predict(X_test), color='red', label = 'regression model')
plt.plot(X_train,lr.predict(X_train), color='red', label = 'regression model')
plt.legend()
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
# Slope
m = lr.coef_
```

```python
# y interceps---b
b = lr.intercept_
```
