#### Calculate value of m from scikitlearn library and fix it
#### We are making class for Gradient Descent only for intercept i.e b

```python
m = 61.29
```

```python
class GDRegressor:
    def __init__(self,learning_rate,epochs):
        self.m = 61.29
        self.b = -120
        self.lr=learning_rate
        self.epochs = epochs
    def fit(self,X,y):
        # calculate b using GD
        for i in range(self.epochs):
            loss_slope = -2 * np.sum(y - self.m*X.ravel() - self.b)
            self.b = self.b - self.lr*loss_slope
            print(loss_slope, self.b)
        print(self.b)
```

```python
gd = GDRegressor(0.001,1000)
```

```python
gd.fit(X,y)
```
