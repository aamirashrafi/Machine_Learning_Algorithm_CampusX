## Gradient Descent
## From Scratch

```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# Lets apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
# Lets apply Gradient Descent assuming slope is known and constant = 78.35
# and lets assume the starting value for intercept b = 0
y_pred = ((78.35 * X) + 0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red', label='OLS')
plt.plot(X, y_pred, color = '#00a65a', label='b = 0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0
loss_slop = -2*np.sum(y - m*X.ravel() - b)
loss_slop
```

```python
# Lets take the learning rate = 0.1
lr = 0.1
step_size = loss_slop*lr
b = b - step_size
b
```

```python
y_pred1 = ((m*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X), color='red', label='OLS')
plt.plot(X,y_pred,color='#00a65a',label='b=0')
plt.plot(X,y_pred1,color='#A3E4D7',label='b={}'.format(b))
plt.legend()
plt.show()
```

```python
# for new b we neew to calculate slope
# so slope will be
loss_slop =  -2 * np.sum(y - m*X.ravel()-b)
step_size = lr*loss_slop
b = b-step_size
```

```python
y_pred2 = ((m*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X), color='black', label='OLS')
plt.plot(X,y_pred2,color='#00a65a',label='b={}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label='b={}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b=0')
plt.legend()
plt.show()
```

```python
loss_slop =  -2 * np.sum(y - m*X.ravel()-b)
loss_slop
```

```python
step_size = lr*loss_slop
step_size
```

```python
b = b-step_size
b
```

```python
y_pred3 = ((m*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X), color='black', label='OLS')
plt.plot(X,y_pred3, color='#00a65a',label='b={}'.format(b))
plt.plot(X,y_pred2,color='#00a65a',label='b={}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label='b={}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b=0')
plt.legend()
plt.show()
```

### Gradient Descent From Scratch - Using Loop

```python
b = -100
m = 78.35
lr = 0.01

epochs = 100

for i in range(epochs):
   loss_slope = -2 * np.sum(y - m*X.ravel() - b)
   b = b - (lr*loss_slope)
   y_pred = m * X + b
   plt.plot(X,y_pred)
plt.scatter(X,y)
```
